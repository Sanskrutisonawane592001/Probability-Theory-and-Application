# PLOTTING GRAPH TO SHOW BINOMIAL DISTRIBUTION APPROACHING POISSON DISTRIBUTION AS SAMPLE SZE INCREASES 

N = 10000 ; lam = 14 ; (size)*p = lam ; p = lam/(size)

```python
from numpy import random 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = random.binomial(n = 10000, p = 14/210, size = 210)
sns.distplot(data)
plt.title('Binomial Distribution graph when size = 210')
plt.show()
print(data)
```

```python
data = random.binomial(n = 10000, p = 14/695, size = 695)
sns.distplot(data)
plt.title('Binomial Distribution graph when size = 695')
plt.show()
print(data)
```

```python
data = random.binomial(n = 10000, p = 14/1180, size = 1180)
sns.distplot(data)
plt.title('Binomial Distribution graph when size = 1180')
plt.show()
print(data)
```

```python
data = random.binomial(n = 10000, p = 14/1665, size = 1665)
sns.distplot(data)
plt.title('Binomial Distribution graph when size = 1665')
plt.show()
print(data)
```

```python
data = random.poisson(lam=14 , size = 4100)
sns.distplot(data)
plt.title('poisson Distribution graph when size = 4100')
plt.show()
print(data)
```

```python
sns.distplot(random.binomial(n = 10000, p = 14/210, size = 210))
sns.distplot(random.binomial(n = 10000, p = 14/695, size = 695))
sns.distplot(random.binomial(n = 10000, p = 14/1180, size = 1180))
sns.distplot(random.binomial(n = 10000, p = 14/1665, size = 1665))
sns.distplot(random.poisson(lam = 14, size = 4100))
plt.title('Binomial Distribution and Poisson Distribution graphs stacked together')
plt.show()
```
